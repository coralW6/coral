```python
#20、对缺失值进行填充
df.fillna(10000)
```

```python
#21、对缺失值的行进行删除
#any:该行有一个值为NaN就删掉  all:所有值为NaN就删掉
df.dropna(how='all')
```

```python
#22、对所有值进行bool转化
pd.notnull(df)  #不是None的值置为True,是None的置为False
#pd.isnull(df)   #不是None的值置为False,是None的置为True
```

```python
#23、获取每一列的平均值
#print df.describe()
df.mean()
```

```python
#24、df.std(ddof)
#ddof=0 表示计算总体标准差
#ddof=1 表示计算样本标准差
df.std(1)
```

```python
#25、计算一阶差分 Y(k)=X(k)-X(k-1)
#计算同列该值和上一个值的差值
df.diff()
```

```python
#26、调用函数
df.apply(lambda x : x.max()- x.min())

# def fun(x):
#     return x.max()-x.min()
# df.apply(fun, axis=0)
```

```python
#27、分组并对每个分组执行count函数或者sum()
df.groupby('col_a').sum()
```

```python
#28、group by a and b
df.groupby([date_time, 'col_a']).count()
```

```python
#29、生成数据透视表
pd.pivot_table(df, values='col_c', index=[date_time, 'col_a'], columns='col_b',aggfunc=lambda x:sum(x))
```
