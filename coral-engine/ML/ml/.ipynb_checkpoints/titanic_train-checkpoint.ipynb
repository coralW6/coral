```python
#!/usr/bin/python
# -*- coding:utf-8 -*-
import sys
import os
import pandas as pd  #分析数据
import numpy as np #计算
import sklearn  #算法模型
import matplotlib.pyplot as plt #画图
import seaborn as sns
%matplotlib inline
```

```python
data_train = pd.read_csv('train.csv')
print data_train.info(), '\n'
data_train.describe()
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei'] # 用来正常显示中文标签
plt.rcParams['font.family']='sans-serif'
plt.rcParams['axes.unicode_minus'] = False # 用来正常显示负号
fig = plt.figure()
fig.set(alpha=0.2) # 设定图表颜色alpha参数
```

```python
#plt.subplot2grid((1,1), (0,0)) # 在一张大图里分列几个小图
data_train.Survived.value_counts().plot(kind='bar')# 柱状图
plt.title(u"获救情况 (1为获救)") # 标题
plt.ylabel(u"人数") # Y轴标签
plt.show()
```

```python
#乘客等级分布
data_train.Pclass.value_counts().plot(kind="bar") # 柱状图显示
plt.ylabel(u"人数")
plt.title(u"乘客等级分布")
plt.show()
```

```python
#乘客等级和是否获救的分布
Survived_0 = data_train.Pclass[data_train.Survived == 0].value_counts() # 未获救
Survived_1 = data_train.Pclass[data_train.Survived == 1].value_counts() # 获救
df = pd.DataFrame({u'获救':Survived_1,u'未获救':Survived_0})
df.plot(kind = 'bar', stacked = True)
plt.title(u'各乘客等级的获救情况')
plt.xlabel(u'乘客等级')
plt.ylabel(u'人数')
plt.show()
#结论：Pclass字段对Survived有影响
```

```python
#性别和获救的关系
Survived_m = data_train.Survived[data_train.Sex == 'male'].value_counts()
Survived_f = data_train.Survived[data_train.Sex == 'female'].value_counts()
df = pd.DataFrame({u'男性':Survived_m,u'女性':Survived_f})
df.plot(kind = 'bar', stacked = True)
plt.title(u'按性别看获救情况')
plt.xlabel(u'性别')
plt.ylabel(u'人数')
plt.show()
#结论：未获救的男性占比很大，获救的女性占比很大
```

```python
#船舱等级和女性双重因素下，看获救情况
fig = plt.figure()
plt.title(u'根据舱等级和女性的获救情况')

ax1 = fig.add_subplot(131) # 将图像分为1行4列，从左到右从上到下的第1块
data_train.Survived[data_train.Sex == 'female'][data_train.Pclass == 1].value_counts().plot(kind = 'bar', label = 'female high class', color = '#FA2479')
ax1.set_xticklabels([u'获救',u'未获救'], rotation = 0) # 根据实际填写标签
ax1.legend([u'女性/高级舱'], loc = 'best')

ax2 = fig.add_subplot(132, sharey = ax1)
data_train.Survived[data_train.Sex == 'female'][data_train.Pclass == 2].value_counts().plot(kind = 'bar', label = 'female low class', color = 'pink')
ax2.set_xticklabels([u"未获救", u"获救"], rotation=0)
plt.legend([u"女性/中级舱"], loc='best')

ax2 = fig.add_subplot(133, sharey = ax1)
data_train.Survived[data_train.Sex == 'female'][data_train.Pclass == 3].value_counts().plot(kind = 'bar', label = 'female low class', color = 'pink')
ax2.set_xticklabels([u"未获救", u"获救"], rotation=0)
plt.legend([u"女性/低级舱"], loc='best')
plt.show()

print '高级仓女性人数：', data_train[data_train.Sex == 'female'][data_train.Pclass == 1]['PassengerId'].count() , '\n'
print '中级仓女性人数：', data_train[data_train.Sex == 'female'][data_train.Pclass == 2]['PassengerId'].count() , '\n'
print '低级仓女性人数：', data_train[data_train.Sex == 'female'][data_train.Pclass == 3]['PassengerId'].count() , '\n'

#结论：高级仓女性获救率很高
```

```python
fig = plt.figure()
plt.title(u'根据舱等级和男性的获救情况')

ax3 = fig.add_subplot(131, sharey = ax1)
data_train.Survived[data_train.Sex == 'male'][data_train.Pclass == 1].value_counts().plot(kind = 'bar', label = 'male high class', color = 'lightblue')
ax3.set_xticklabels([u'未获救',u'获救'], rotation = 0)
plt.legend([u'男性/高级舱'], loc = 'best')

ax3 = fig.add_subplot(132, sharey = ax1)
data_train.Survived[data_train.Sex == 'male'][data_train.Pclass == 2].value_counts().plot(kind = 'bar', label = 'male high class', color = 'lightblue')
ax3.set_xticklabels([u'未获救',u'获救'], rotation = 0)
plt.legend([u'男性/中级舱'], loc = 'best')

ax4 = fig.add_subplot(133, sharey = ax1)
data_train.Survived[data_train.Sex == 'male'][data_train.Pclass == 3].value_counts().plot(kind = 'bar', label = 'male low class', color = 'steelblue')
ax4.set_xticklabels([u'未获救',u'获救'], rotation = 0)
plt.legend([u'男性/低级舱'], loc = 'best')
plt.show()

print '高级仓男性人数：', data_train[data_train.Sex == 'male'][data_train.Pclass == 1]['PassengerId'].count() , '\n'
print '中级仓男性人数：', data_train[data_train.Sex == 'male'][data_train.Pclass == 2]['PassengerId'].count() , '\n'
print '低级仓男性人数：', data_train[data_train.Sex == 'male'][data_train.Pclass == 3]['PassengerId'].count() , '\n'

#低级仓的男生获救率很低
```

```python
#分析年龄和获救的关系
sns.set_style('whitegrid')
fig, axis1 = plt.subplots(1,1,figsize=(50,10))
df = data_train[["Survived", "Age"]].groupby(['Age'],as_index=False).mean()
sns.barplot(x='Age', y='Survived', data=df)
#老人和小孩的获救率较高
```

```python
#导入测试集数据  将两个集合的数据做处理
data_test = pd.read_csv('test.csv') # 导入测验集数据
print data_test.info()

y = data_train['Survived'] # 将训练集Survived 数据存储在y中
print y

del data_train['Survived'] # 删除训练集Survived数据
```

```python
PassengerId = data_test['PassengerId'] # 存储测试集乘客ID
#print PassengerId

df = pd.merge(data_train, data_test,how='outer') # 合并无Survived数据的训练集与测验集，how = ‘outer’ 意为并集
#print df

#删掉无关因素
#df = df.drop(['Name','PassengerId','Ticket','Cabin'],axis=1) # 删除姓名、ID、船票信息、客舱信息，axis=0 删除行，=1 删除列
df = df.drop(['Name','PassengerId','Ticket'],axis=1) # 删除姓名、ID、船票信息，axis=0 删除行，=1 删除列
print df
```

```python
#补全特征值：

df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
#1、年龄，用平均值补全
df['Age'] = df['Age'].fillna(df['Age'].mean())  
#2、年龄，随机森林算法补全
# from sklearn.ensemble import RandomForestRegressor
# age_df = df[['Age','Fare', 'Parch', 'SibSp', 'Pclass']]
# known_age = age_df[age_df.Age.notnull()].as_matrix()
# unknown_age = age_df[age_df.Age.isnull()].as_matrix()
# y = known_age[:, 0]  # y即目标年龄
# X = known_age[:, 1:] # X即特征属性值
# #print X.tolist()
# # fit到RandomForestRegressor之中
# rfr = RandomForestRegressor(random_state=0, n_estimators=2000, n_jobs=-1)
# rfr.fit(X, y)

# #用得到的模型进行未知年龄结果预测
# predictedAges = rfr.predict(unknown_age[:, 1::])
# print predictedAges
# df.loc[(df.Age.isnull()), 'Age'] = predictedAges
# #df['Age'].fillna(predictedAges)
# print df['Age']

#补全船舱信息
df['Cabin'] = pd.notnull(df['Cabin'])
print df['Cabin']
    
#港口 用出现次数最多的港口补全
#print data_train['Embarked'].value_counts().index[0]
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].value_counts().index[0])
#将性别和港口用哑变量表示 （将不能够定量处理的变量量化，根据这些因素的属性类型，构造只取“0”或“1”的人工变量）
dumm = pd.get_dummies(df[['Sex','Embarked']]) # '哑变量'矩阵
df = df.join(dumm) #将哑变量矩阵和原有的矩阵合并
# print df
del df['Sex']
del df['Embarked']
print df
```

```python
#对age和fare进行数据降维
df['Age'] = (df['Age']-df['Age'].min()) /(df['Age'].max()-df['Age'].min())
df['Fare'] = (df['Fare']-df['Fare'].min()) /(df['Fare'].max()-df['Fare'].min())
#print df['Fare']
# 不考虑SibSp Parch 试试
# del df['SibSp']
# del df['Parch']
# del df['Fare']
#特征值 [Pclass, Age, SibSp, Parch, Fare, Sex_female, Sex_male, Embarked_C, Embarked_Q, Embarked_S]

data_train = df[:len(data_train)]
data_test = df[len(data_train):]

print data_train
data_train.to_csv('debug.csv')

```

```python
from sklearn.cross_validation import train_test_split
# 以7：3（0.3）将训练集与获救结果随机拆分，随机种子为42 用于训练和验证
X_train, X_val, y_train, y_val = train_test_split(data_train,y,test_size=0.3, random_state=42)

from sklearn.linear_model import LogisticRegression # 引入逻辑回归
LR = LogisticRegression()
#LR.fit(data_train, y)
LR.fit(X_train, y_train) # 训练数据
print LR
print '训练集准确率：\n',LR.score(X_train, y_train) # 分数
print '验证集准确率：\n',LR.score(X_val, y_val)

#预测测验集
print data_test
#data_test.to_csv('tmp.csv')
pred = LR.predict(data_test) # pred 为预测结果
print pred
pred = pd.DataFrame({'PassengerId':PassengerId.values, 'Survived':pred}) # 格式化预测结果
print pred
pred.to_csv('pred_LR01.csv',index=None) # 导出数据
```
